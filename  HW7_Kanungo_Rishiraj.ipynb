```r
# create vectors
id <- c(1, 2, 3, 4, 5)
name <- c("Peter", "Amy", "Ryan", "Gary", "Michelle")
salary <- c(623.30, 514.20, 611.00, 729.00, 843.25)
```

```r
# create dataframe from vectors
df <- data.frame(id, name, salary)
df
```

```r
# create a column of "department" and assign names to it
department <- c("IT", "Finance", "Computer Science", "Security", "Human Resources")
cbind(df, department)
```

```r
# Extract 1st, 3rd, and 5th rows, with 2nd and 3rd columns, using the c function
df[c(1,3,5), 2:3]
```

```r
# Plot a bar chart showing the salary of row 1,4,5 and label the bar with peopleâ€™s name correspondingly
bar_y <- df[c(1,4,5), 3]
bar_x <- df[c(1,4,5), 2]
barplot(bar_y,names.arg = bar_x)
```

```r
# plot a pie chart, based on this data framework, only showing the highest salary, lowest salary, and
# median salary, give your plot a title, and label each pie.
pie_x <- df[,3]
pie_x
pie_y <- df[,2]
pie_y
colors <- c("blue", "yellow", "cyan", "red", "white")
pie(pie_x, labels = pie_y, main = "Pie Chart", col = colors)
```

```r
open_file <- function(){
    
    while(TRUE){
        user_input <- readline(prompt="Enter file name: ")
        
        output <- tryCatch(
        {
            my_data <- read.delim(user_input)
            print(my_data)
            break
        },
        error=function(cond) {
            print("Error")
            # Choose a return value in case of error
            return(NA)
        }
    )
    }    
}
```

```r
open_file()
```
