```python
# import dependencies
import numpy as np
import matplotlib.pyplot as plt
import json
import pandas as pd
%matplotlib inline
```

```python
# Question 1 (6 points): Counting the frequencies in a list using dictionary in Python
# Given an unsorted list of some elements:

# Find the frequency of each distinct element in the list using a dictionary.
def countFreq(my_list):
    freq = {}
    for element in my_list:
        if (element in freq):
            freq[element] += 1
        else:
            freq[element] = 1
 
    for k, v in freq.items():
        print ("% d : % d"%(k, v))
        
    return freq
       
# Plot the frequency results, e.g. using histogram
# TODO: FIX PLOT
def plotFreq(my_dict):
    plt.hist(my_dict.values(), bins=4)  # density=False would make counts
    plt.ylabel('Frequency')
    plt.xlabel('Data');
    
# Store this dictionary results into json file and store in your local computer.
def saveJSON(my_dict):
    with open("sample.json", "w") as outfile:
        json.dump(my_dict,outfile)
```

```python
my_list = [2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]

freq = countFreq(my_list)

plotFreq(freq)

saveJSON(freq)
```

```python
# Choice 2: How Much Have You Spent on Amazon?

# create dataframe
df = pd.read_csv('amazon_orders.csv')
df.head() # peek the head of the DF
```

```python
# data preprocessing
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float) # convert the string values into floats
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float) # convert the string values into floats
df['Order Date'] = pd.to_datetime(df['Order Date']) # convert the strings into datetime obj.
df.head()
```

```python
# Quantative metrics
average_spending = df["Total Charged"].mean()
median_spending = df["Total Charged"].median()
most_expensive_order = df["Total Charged"].max()
least_expensive_order = df["Total Charged"].min()


print("The average amount per order I make is %5.2f"%average_spending)
print("The mediam spending that I do on Amazon is %5.2f"%median_spending)
print("My most expensive order is $%5.2f"%most_expensive_order)
print("My least expensive order is $%5.2f"%least_expensive_order)

# median spend is quite a bit lower than my average spend. Making a few expensive orders.
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
# group the orders that happen the same day together instead as individual orders
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
# although the biggest single order was just over $200, there was a single day in July where several orders totaled up to well over $400.

daily_orders.plot.bar(figsize=(20,10))
```
