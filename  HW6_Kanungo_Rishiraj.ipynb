```r
library(readr)
mydata <- read.csv("amazon_orders.csv")
df <- data.frame(mydata)

# preprocess data
df$Total.Charged <- as.numeric(sub("\\$","", df$Total.Charged))
df$Total.Charged
df$Tax.Charged <- as.numeric(sub("\\$","", df$Tax.Charged))
df$Tax.Charged
df$Order.Date <- as.Date(df$Order.Date, "%m/%d/%Y")
df$Order.Date
```

```r
average_spending <- mean(df$Total.Charged)
cat("The average amount per order I make is ", format(round(average_spending, 2), nsmall = 2), "\n")
median_spending <- median(df$Total.Charged)
cat("The mediam spending that I do on Amazon is", format(round(median_spending, 2), nsmall = 2), "\n")
most_expensive_order <- max(df$Total.Charged)
cat("My most expensive order is", format(round(most_expensive_order, 2), nsmall = 2), "\n")
least_expensive_order <- min(df$Total.Charged)
cat("My least expensive order is", format(round(least_expensive_order, 2), nsmall = 2), "\n")
```

```r
daily_orders <- aggregate(df["Total.Charged"], by=df["Order.Date"], sum)
daily_orders
```

```r
barplot(daily_orders$Total.Charged, names.arg = daily_orders$Order.Date)
```
